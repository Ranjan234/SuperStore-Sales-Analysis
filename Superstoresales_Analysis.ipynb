# **Superstore Sales** 🛒🏷️🛍️📦🏪
  - ## **About Dataset**
 - ## *"Superstore Sales" generally refers to the sales data and performance of a superstore or a large retail store that offers a wide range of products and merchandise. These superstores are typically known for their extensive inventory, competitive pricing, and one-stop shopping experience.*

- ## *Analyzing superstore sales data is essential for store owners, managers, and analysts to understand the store's performance, identify trends, and make informed business decisions. Here are some key aspects of superstore sales analysis:*

- ### **Sales Revenue: Tracking the total sales revenue over time provides insights into the overall financial performance of the superstore. It helps identify peak seasons and periods of high or low sales.**

- ### **Sales by Product Category: Analyzing sales by product category (e.g., electronics, apparel, home goods) helps understand which categories are driving revenue and which may require improvement.**

- ### **Sales by Region: Superstores often have multiple locations. Analyzing sales by region helps identify the best-performing stores and areas for expansion or improvement.**

- ### **Sales by Customer Segments: Understanding sales patterns among different customer segments can inform targeted marketing strategies and customer engagement.**

- ### **Sales Trends: Identifying sales trends over time, such as monthly or seasonal variations, can help in inventory management and marketing planning.**

- ### **Product Performance: Analyzing individual product sales can help identify popular products, slow-moving items, and potential stockouts.**

- ### **Customer Behavior: Studying customer behavior, such as purchase frequency, average transaction value, and repeat purchases, provides insights into customer preferences and loyalty.**

- ### **Promotions and Discounts: Evaluating the impact of promotions and discounts on sales can help optimize marketing strategies.**

- ### **Sales Forecasting: Using historical sales data, superstore owners and managers can forecast future sales, plan inventory levels, and make data-driven decisions.**

 ### **2. Handel Imbalances  & mIssing values**

```python
# Postal code has null values
# Convert float postal codes to string
df['Postal_Code'] = df['Postal_Code'].astype(str)
```

```python
# Clean up the .0 that came from flaot conversion
df['Postal_Code'] = df['Postal_Code'].str.replace('.0', '')
```

```python
# Find the most common postal code
df.fillna({"Postal_Code": df['Postal_Code'].mode()[0]}, inplace=True)
df.Postal_Code.unique()
```

### **3. Change date columns data type**

```python
# Change data type order_id, ship_date to date_tme
df["Ship_Date"] = pd.to_datetime(df["Ship_Date"], format='%d/%m/%Y')
df["Order_Date"] = pd.to_datetime(df["Order_Date"], format='%d/%m/%Y')
df.head()
```

```python
df.info()
```

 - ## **4. Create New Features**

```python
# Create total days of shipment , new features
df['Days_to_Ship'] = (df['Ship_Date'] - df['Order_Date']).dt.days
df['Days_to_Ship'].std()
df['Days_to_Ship'].describe()
```

```python
df.head()
```

```python
# Calculate average unit price per Product_ID
product_price_map = df.groupby('Product_ID')['Sales'].mean().to_dict()

# Map the unit_price column based on Product_ID
df['unit_price'] = df['Product_ID'].map(product_price_map)

# Display the results
display(df[['Product_ID', 'Category', 'unit_price', 'Sales']].head(10))
```

```python
# Create Quantity column
df['Quantity'] = (df['Sales'] / df['unit_price']).round(0).astype(int)
```

```python
# Create profit column
df['Profit'] = df['Sales'] - (df['Quantity'] * df['unit_price'])
```

```python
# Create profit margin
df['Profit_Margin'] = ((df['Profit'] / df['Sales']) * 100).round(2)
```

 ## **STEP 5 : Visualization & Summarization**


```python
# Check outliers
Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)
```

```python
IQR  = Q3 - Q1
```

```python

Lower_fence = Q1 - 1.5 * IQR # ← very low sales
Upper_fence = Q3 + 1.5 * IQR # ← very high sales
print(Lower_fence)
print(Upper_fence)
```

```python
outliers = df[(df['Sales'] < (q1 - 1.5 * IQR)) | 
              (df['Sales'] > (q3 + 1.5 * IQR))]
```

```python
 
# Step 1: Check outlier statistics
print("=== OUTLIER SUMMARY ===")
print(f"Total outliers: {len(outliers)}")
print(f"% of data: {len(outliers)/len(df)*100:.2f}%")
print(f"\nSales range in outliers:")
print(f"  Min: ${outliers['Sales'].min():,.2f}")
print(f"  Max: ${outliers['Sales'].max():,.2f}")
print(f"  Mean: ${outliers['Sales'].mean():,.2f}")
print(f"\nProfit in outliers:")
print(f"  Total Profit: ${outliers['Profit'].sum():,.2f}")
```

```python
#  Total revenue
Total_Revenue = df['Sales'].sum().round(2)
print(f'Total Revenue: {Total_Revenue}')
```

```python
# Revenue by category
Revenue_by_Category = df.groupby('Category')['Sales'].sum().round(2).sort_values(ascending=False)
print(Revenue_by_Category)
```

```python
# Revenue by Region
Revenue_by_Region = df.groupby('Region')['Sales'].sum().round(2).sort_values(ascending=False)
print(Revenue_by_Region)
```

```python
# Revenue by Sub_Category
Revenue_by_Sub_Category = df.groupby('Sub_Category')['Sales'].sum().round(2).sort_values(ascending=False)
print(Revenue_by_Sub_Category)
```

```python
# Calculate average price
avg_price = df['Sales'].mean().round(2)
print(f'Average Price: {avg_price}')
```

```python
# Average price per category
avg_price_by_category = df.groupby('Category')['Sales'].mean().round(2)
print(avg_price_by_category)
```

```python
# Quantity sold per category
quantity_sold_by_category = df.groupby('Category')['Quantity'].sum()
print(quantity_sold_by_category)
```

```python
# Quantity sold per region
quantity_sold_by_region = df.groupby('Region')['Quantity'].sum()
print(quantity_sold_by_region)
```

```python
# Quantity sold per subcategory
quantity_sold_by_subcategory = df.groupby('Sub_Category')['Quantity'].sum()
print(quantity_sold_by_subcategory)
```

```python
# Profit by segment
profit_by_segment = df.groupby('Segment')['Profit'].sum().round(2)
print(profit_by_segment)
```
